Feature Scaling

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df['duration_minutes'] = df['duration'].str.extract('(\d+)').astype(float)

numerical_cols = ['release_year', 'duration_minutes']
df_numeric = df[numerical_cols].dropna()

std_scaler = StandardScaler()
df_numeric_std = std_scaler.fit_transform(df_numeric)

minmax_scaler = MinMaxScaler()
df_numeric_minmax = minmax_scaler.fit_transform(df_numeric)

df_scaled = pd.DataFrame(df_numeric_std, columns=[f"{col}_std" for col in numerical_cols])
df_scaled[[f"{col}_minmax" for col in numerical_cols]] = pd.DataFrame(df_numeric_minmax)

df_scaled.head()
```

Feature scaling ensures that all features have a similar range, improving the performance of machine learning models.

It helps to

(1) Speed up convergence

(2) Prevent dominance

(3) Ensure fair weight assignment

Clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_numeric_std)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch


# Generate Dendrogram
plt.figure(figsize=(10, 5))
dendrogram = sch.dendrogram(sch.linkage(df_numeric_std, method="ward"))
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

How many clusters are there

(1)Based on the results from the K-Means clustering algorithm, the Elbow Method suggests that the best number of clusters is 3. This is because the "elbow point" (where the curve bends) clearly indicates that 3 clusters provide the best balance.

(2)Similarly, using Hierarchical Clustering with a Dendrogram, we also see that the ideal number of clusters is 3. This is because the vertical separation is highest at this point, and the colored-clusters confirms this.
